```python
# Asignatura: Computación Gráfica e Interacción Humano-Computadora
# Autor: Téllez González Jorge Luis
# Descripción: Implementación del algoritmo de dibujado de líneas Bresenham.

import matplotlib.pyplot as plt
import numpy as np
import math

print("Algoritmo Bresenham para el dibujado de líneas. \nIntroduzca a continuación las coordenadas de",
      "inicio y final de la línea a dibujar:")

x1 = int(input("X1: "))
y1 = int(input("Y1: "))

x2 = int(input("X2: "))
y2 = int(input("Y2: "))

if(x2 == x1 and y2 == y1):
    print("Valores iguales!")
else:
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    p0 = ((2*dy)-dx)
    dosdy = 2 * dy
    dosdx = 2 * dx
    deltaxydoble = dosdy - dosdx
    
    if (x1 > x2):
        x = x2
        y= y2
        x2 = x1
    else:
        x = x1
        y = y1
    
    xline = []
    yline = []
    xline.append(x)
    yline.append(y)
    
    while (x < x2):
        x+=1
        if (p0 < 0):
            p0+= dosdy
        else:
            y+=1
            p0+= deltaxydoble
        xline.append(x)
        yline.append(y)
        

fig1 = plt.figure(figsize=(8, 6))
ax = fig1.gca()
plt.xlim(0, max(xline)+2)
plt.ylim(0, max(yline)+2)
ax.set_xticks(np.arange(0, max(xline)+2, 1))
ax.set_xticks(np.arange(0, max(xline)+2, 1)-0.5, minor=True)

ax.set_yticks(np.arange(0, max(yline)+2, 1))
ax.set_yticks(np.arange(0, max(yline)+2, 1)-0.5, minor=True)

plt.grid(True, which='minor')
plt.title("Muestro discretizado de la línea")
ax.plot(xline, yline, linestyle='--', marker='o',
        color='r', label='Puntos calculados')

for f, b in zip(xline, yline):
    rectangle = plt.Rectangle((f-0.5, b-0.5), 1, 1, fc='blue', ec="red")
    plt.gca().add_patch(rectangle)
plt.margins(0)
plt.show()

        
plt.title('Algoritmo de Bresenham')
plt.xlabel('Eje X')
plt.xlabel('Eje Y')
plt.grid(True)
plt.plot(xline, yline, linestyle='--', marker='o', color='r', label='Puntos calculados')
plt.legend()

print(xline)
print(yline)
```
