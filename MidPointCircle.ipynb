```python
# Asignatura: Computaci√≥n Gr√°fica e Interacci√≥n Humano-Computadora
# Autor: T√©llez Gonz√°lez Jorge Luis
# Descripci√≥n: Implementaci√≥n del algoritmo de dibujado de circunferencias.

import matplotlib.pyplot as plt
import numpy as np


xline = []
yline = []

def setPixel(xcord, ycord):
    xline.append(xcord)
    yline.append(ycord)

def circleMidpoint(xc, yc, radio):
    global x 
    global y
    x = radio
    y = 0
    
    print(x,y)
    #setPixel(radio + 0, 0 + 0)  [10,0]
    setPixel(x + xc, y +yc)
    #setPixel(0 + 0, 0 + 0)  [0,10]
    setPixel(y + xc, x + yc)
    
    #setPixel(radio + 0, 0 + 0)  [-10,0]
    setPixel(-x + xc, y + yc)
    #setPixel(0 + 0, 0 + 0)  [0,-10]
    setPixel(y + xc, -x + yc)
    
    p = 1 - radio 
    
    while (x > y):
        y += 1
        if (p <= 0):
            p += 2 * y + 1
        else:
            x -= 1
            p += 2 * (y-x) + 1
        
        if (x < y):
            break
        
        setPixel(xc + x, yc + y)
        setPixel(xc - x, yc + y)
        setPixel(xc + x, yc - y)
        setPixel(xc - x, yc - y)
        
        if x!=y:
            setPixel(xc + y, yc + x)
            setPixel(xc - y, yc + x)
            setPixel(xc + y, yc - x)
            setPixel(xc - y, yc - x)

print("Algoritmo del punto medio para el dibujado de circunferencias. \nIntroduzca a continuaci√≥n el radio",
    "y el punto central de la circunferencia a dibujar:")

r = int(input("Radio: "))
x1 = int(input("X1: "))
y1 = int(input("Y1: "))

circleMidpoint(x1, y1, r)

print(xline)
print(yline)

fig1 = plt.figure(figsize=(8, 6))
ax = fig1.gca()
plt.xlim(-max(xline)-2, max(xline)+2)
plt.ylim(-max(yline)-2, max(yline)+2)
ax.set_xticks(np.arange((max(xline)+2)*-1, max(xline)+2, 1))
ax.set_xticks(np.arange((max(xline)+2)*-1, max(xline)+2, 1)-0.5, minor = True)

ax.set_yticks(np.arange((max(yline)+2)*-1, max(yline)+2, 1))
ax.set_yticks(np.arange((max(yline)+2)*-1, max(yline)+2, 1)-0.5, minor=True)

plt.grid(True, which= 'minor')
plt.title("Muestreo discretizado de la l√≠nea")
ax.plot(xline, yline, linestyle='--', marker='o',
        color='r', label='Puntos calculados')

for f, b in zip(xline, yline):
    rectangle = plt.Rectangle((f-0.5, b-0.5), 1, 1, fc='blue', ec="red")
    plt.gca().add_patch(rectangle)
plt.margins(0)
plt.show()

plt.title('Algoritmo del Punto Medio')
plt.xlabel('Eje X')
plt.xlabel('Eje Y')
plt.grid(True)
plt.plot(xline, yline,  'ro', color='r')
plt.legend()
```
