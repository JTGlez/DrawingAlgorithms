```python
# Asignatura: Computación Gráfica e Interacción Humano-Computadora
# Autor: Téllez González Jorge Luis
# Descripción: Implementación del algoritmo de dibujado de líneas DDA.

import matplotlib.pyplot as plt
import numpy as np


def listOfTuples(l1, l2):
    return list(map(lambda x, y: (x, y), l1, l2))

print("Algoritmo DDA para el dibujado de líneas. \nIntroduzca a continuación las coordenadas de",
    "inicio y final de la línea a dibujar:")

x1 = int(input("X1: "))
y1 = int(input("Y1: "))

x2 = int(input("X2: "))
y2 = int(input("Y2: "))

if(x2 == x1 and y2 == y1):
    print("Valores iguales!")
else:
    # Cálculo de los diferenciales.
    dx = int(x2 - x1)
    dy = int(y2 - y1)

    # Determinación del mayor diferencial.
    if abs(dx) > abs(dy):
        steps = abs(dx)
    else:
        steps = abs(dy)

    # Cálculo de los incrementos
    xinc = dx/steps
    yinc = dy/steps

    # Listas con los valores de X y Y en cada paso.
    xline = []
    yline = []
    xline.append(x1)
    yline.append(y1)

    # Iterador
    i = 0

    while i < steps:
        i += 1
        x1 = x1 + xinc
        y1 = y1 + yinc
        xline.append(round(x1))
        yline.append(round(y1))

fig1 = plt.figure(figsize=(8, 6))
ax = fig1.gca()
plt.xlim(0, max(xline)+2)
plt.ylim(0, max(yline)+2)
ax.set_xticks(np.arange(0, max(xline)+2, 1))
ax.set_xticks(np.arange(0, max(xline)+2, 1)-0.5, minor=True)

ax.set_yticks(np.arange(0, max(yline)+2, 1))
ax.set_yticks(np.arange(0, max(yline)+2, 1)-0.5, minor=True)

plt.grid(True, which='minor')
plt.title("Muestreo discretizado de la línea")
ax.plot(xline, yline, linestyle='--', marker='o',
        color='r', label='Puntos calculados')

for f, b in zip(xline, yline):
    rectangle = plt.Rectangle((f-0.5, b-0.5), 1, 1, fc='blue', ec="red")
    plt.gca().add_patch(rectangle)
plt.margins(0)
plt.show()

plt.title('Algoritmo DDA')
plt.xlabel('Eje X')
plt.xlabel('Eje Y')
plt.grid(True)
plt.plot(xline, yline, linestyle='--', marker='o', color='r', label ='Puntos calculados')
plt.legend()

print(xline)
print(yline)
```
